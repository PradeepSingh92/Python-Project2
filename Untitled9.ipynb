```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.w3schools.com"
```

```python
#step1 - get the html
r = requests.get(url)
htmlcontent = r.content
print(htmlcontent)
```

```python
#step2 parse the html (creating soup)
soup = BeautifulSoup(htmlcontent,'html.parser')
print(soup)
```

```python
print(soup.prettify)
```

```python
#step3 HTML tree Traversal - contant like body head tags style ..etc 
title = soup.title
print(title)
```

```python
print(type(title))
```

```python
print(type(title.string))
```

```python
# commonly used types of / find objects 
# tag print(type(title)
# navigablestring print(type(soup))
#Beautifulsoup print(type(title.string))
#comment
#get the title of the html page 
#title=soup.title

#get all the paragraphs from the page 
paras = soup.find_all('p')
print(paras)
```

```python
#get all the anchor tags  from the page 
anchors = soup.find_all('a')
print(anchors)
```

```python
#get class of any element in the html page 
print(soup.find('p'))
```

```python
#get classes of any element in the html page
print(soup.find('p')['class'])
```

```python
# find all te elements with classlead
print(soup.find_all("p", class_="Lead"))
```

```python
# get the text from the elements tags/soup
print(soup.find("p").get_text())
print(soup.get_text())
```

```python
#get all the links on the page 
for link in anchors:
    print(link.get('href'))
```

```python
# get all links on the page :
all_links = set()
for link in anchors:
    if(link.get('href') != '#'):
        linkText = "https://www.w3schools.com" +link.get('href')
        all_links.add(link)
        print(linkText)
```
